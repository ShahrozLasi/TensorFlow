````
rank1_tensor = tf.Variable(['Test'], tf.string)
rank2_tensor = tf.Variable([['test', 'ok'], ['test', 'yes']], tf.string)

tf.rank(rank1_tensor),tf.rank(rank2_tensor)
````

````
tensor1 = tf.ones([1,2,3]) ## tf.ones() creates a shape [1,2,3] tensor full of ones
# 1 bigger list, 2 list within the bigger list with 3 elements in each of the 2 smaller lists.
tensor2 = tf.reshape(tensor1, [2,3,1]) #reshaping exisitng data to shapee [2,3,1]
# 2 big list, 3 small list in each  with 1 element.
tensor3 = tf.reshape(tensor2, [3,-1]) #-1 tell the tensor to calculate the size of the dimension in that place 
                                      # this will reshape the tensor to [3,2]
# The number of elements in the reshaped tensor MUST match the number in the original

print(tensor1)
print(tensor2)
print(tensor3)  
````
